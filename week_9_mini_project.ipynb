## Establish Datset Connection

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the wine dataset from UCI repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
column_names = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash",
                 "Magnesium","Total phenols", "Flavanoids",
                 "Nonflavanoid phenols", "Proanthocyanins",
                 "Color intensity", "Hue", "OD280/OD315 of diluted wines",
                 "Proline"]
wine_data = pd.read_csv(url, names=column_names)

assert len(wine_data.columns) == 14

# Display the first few rows of the dataset
wine_data.head()
```

## Demonstrate Data Manipulation

```python
# Filter wines with Alcohol content greater than 14
high_alcohol = wine_data[wine_data['Alcohol'] > 14]

assert len(high_alcohol) == 22

high_alcohol.head()
```

```python
# Average properties of wine by Class
average_properties = wine_data.groupby('Class').mean()
average_properties
```

## Data Visualization

### A a histogram of Alcohol content:

```python
# Plot a histogram of Alcohol content
wine_data['Alcohol'].hist(bins=30, edgecolor='black')
plt.xlabel('Alcohol')
plt.ylabel('Count')
plt.title('Distribution of Alcohol Content')
plt.show()
```

### Bar Plot of Average Alcohol Content by Wine Class:

```python
# Group by wine class and calculate the mean
class_alcohol_avg = wine_data.groupby('Class')['Alcohol'].mean()

# Plotting
class_alcohol_avg.plot(kind='bar', color='skyblue', edgecolor='black')
plt.ylabel('Average Alcohol Content')
plt.title('Average Alcohol Content by Wine Class')
plt.xticks(rotation=0)
plt.show()
```

### Scatter Plot of Alcohol vs. Color Intensity:

```python
plt.figure(figsize=(10,6))
plt.scatter(wine_data['Alcohol'], wine_data['Color intensity'], c=wine_data['Class'],
            cmap='viridis', alpha=0.7, edgecolors='w', linewidth=0.5)
plt.xlabel('Alcohol')
plt.ylabel('Color Intensity')
plt.title('Scatter Plot of Alcohol vs. Color Intensity')
plt.colorbar(label='Wine Class')
plt.show()
```
